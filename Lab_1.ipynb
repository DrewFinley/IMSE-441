```python
import math
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/IMSE 441/Lab_1_1.csv', sep=',', na_values=".", encoding='cp1252')
df 
```

```python
plt.figure(figsize = (70,30))
plt.plot(df['Film'], df['Total Gross in K'], color = 'green', linewidth = 5, linestyle = '-', label = 'Total Gross')
plt.plot(df['Film'], df['Number of cinemas'], color = 'red', linewidth = 5, linestyle = '-.', label = 'Number of Cinemas')
plt.xlabel('Film', fontsize = 35)
plt.ylabel('Gross $ (Thousands of Dollars)', fontsize = 35)
plt.title('Film vs Gross $', fontsize = 50)
plt.legend(fontsize = 75)
plt.show
plt.yticks(fontsize = 20)
```

```python
plt.figure(figsize = (65,20))
plt.bar(df['Film'], df['Weekend in K'], label = " Weekend Gross", alpha = .7)
plt.bar(df['Film'], df['Weeks on release'], label = "Weeks in Theater", alpha = .7)
plt.xlabel('Film', fontsize = 35)
plt.ylabel('$ (Thousands of Dollars)', fontsize = 35)
plt.title('Weekend Gross vs Weeks in Theater', fontsize = 50)
plt.legend(fontsize = 75)
plt.show
plt.yticks(fontsize = 20)
```

```python
Disney = (df[df['Distributor'] =='Disney']['Total Gross in K'])
Universal = (df[df['Distributor'] =='Universal']['Total Gross in K'])
Sony = (df[df['Distributor'] =='Sony Pictures']['Total Gross in K'])
Ahimsa = (df[df['Distributor'] =='Ahimsa Entertainment']['Total Gross in K'])
DG = (df[df['Distributor'] =='DG Tech']['Total Gross in K'])
Dreamz = (df[df['Distributor'] =='Dreamz Entertainment']['Total Gross in K'])
BFI = (df[df['Distributor'] =='BFI']['Total Gross in K'])
```

```python
data = [Disney, Universal, Sony, Ahimsa, DG, Dreamz, BFI]
data
```

```python
plt.figure(figsize=(15,15))
plt.boxplot(data)
plt.title('Distributor vs Gross Income (in thousands of dollars')
plt.xlabel('Distributor', fontsize = 25)
plt.ylabel("Gross Income (thousands of $", fontsize = 25)
plt.xticks([1, 2, 3, 4, 5, 6, 7], ['Disney', 'Universal', 'Sony', 'Ahimsa', 'DG Tech', 'Dreamz', 'BFI'])
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.boxplot([Ahimsa,BFI,Dreamz, DG])
plt.title('Distributor vs Gross Income (in thousands of dollars')
plt.xlabel('Distributor', fontsize = 25)
plt.ylabel("Gross Income (thousands of $", fontsize = 25)
plt.xticks([1, 2, 3, 4], ['Ahimsa', 'BFI', 'Dreamz', 'DG Tech'])
plt.show()
```

```python
Distributor =['Disney', 'Universal', 'Sony', 'Ahimsa', 'DG', 'Dreamz', 'BFI']
Profit = [sum(Disney), sum(Universal), sum(Sony), sum(Ahimsa), sum(DG), sum(Dreamz), sum(BFI)]
plt.figure(figsize= (18,18))
plt.pie(Profit, labels = Distributor, autopct = '%0.2f%%')
plt.title("Distributor's Gross Income", fontsize = 35)
pie = plt.show()
```
